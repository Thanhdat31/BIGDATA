```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("import/Elon 2020-2021.csv")

# Loại bỏ các cột không cần thiết
df = df.drop(columns=['URL', 'Media','Hashtags'])

# Loại bỏ "+00:00" từ cột "Date" bằng cách cắt chuỗi
df['Date'] = df['Date'].str.split('+').str[0]

# Loại bỏ các dòng trùng lặp
df.drop_duplicates(inplace=True)
```

```python
#Kiểm tra các dòng có giá trị thiếu trong một số cột cụ thể 
missing_data = df[['Date', 'Tweet', 'Likes', 'Retweets', 'Quote', 'Replies']].isnull().any(axis=1)

# Lọc các dòng có giá trị thiếu
missing_rows = df[missing_data]

# Hiển thị các dòng có giá trị thiếu
print(missing_rows)
```

```python
#lưu file
df.to_csv('data_cleaned1.csv', index=False) 
```
