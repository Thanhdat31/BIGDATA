```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("import/Elon 2020-2021.csv")

# Lo·∫°i b·ªè c√°c c·ªôt kh√¥ng c·∫ßn thi·∫øt
df = df.drop(columns=['URL', 'Media','Hashtags'])

# Lo·∫°i b·ªè "+00:00" t·ª´ c·ªôt "Date" b·∫±ng c√°ch c·∫Øt chu·ªói
df['Date'] = df['Date'].str.split('+').str[0]

# Lo·∫°i b·ªè c√°c d√≤ng tr√πng l·∫∑p
df.drop_duplicates(inplace=True)
```

```python
#Ki·ªÉm tra c√°c d√≤ng c√≥ gi√° tr·ªã thi·∫øu trong m·ªôt s·ªë c·ªôt c·ª• th·ªÉ 
missing_data = df[['Date', 'Tweet', 'Likes', 'Retweets', 'Quote', 'Replies']].isnull().any(axis=1)

# L·ªçc c√°c d√≤ng c√≥ gi√° tr·ªã thi·∫øu
missing_rows = df[missing_data]

# Hi·ªÉn th·ªã c√°c d√≤ng c√≥ gi√° tr·ªã thi·∫øu
print(missing_rows)
```

```python
#l∆∞u file
df.to_csv('data_cleaned1.csv', index=False) 
```
